```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

# Funções para facilitar a análise


```python
def getData(dado, base):
    df = pd.DataFrame()
    
    df['african_american'] = pd.read_csv('./african-american_' + base + '.csv')[dado]
    df['asian'] = pd.read_csv('./asian_' + base + '.csv')[dado]
    df['caucasian'] = pd.read_csv('./caucasian_' + base + '.csv')[dado]
    #df['hispanic'] = pd.read_csv('./hispanic_' + base + '.csv')[dado] ## RETIREI pois 20% de val não foi suficiente
    df['mix'] = pd.read_csv('./nova_multiracial_' + base + '.csv')[dado]
    return df

    
```

## Resultados


### 1.VGG19

```python
VGG19_acc = getData('accuracy', 'VGG19').rolling(15).mean()
VGG19_vacc = getData('val_accuracy', 'VGG19').rolling(15).mean()


fig, axs = plt.subplots(1, 2, figsize=(15,5), sharey=True)

VGG19_acc.plot(ax=axs[0], title='Acurácia da estutura VGG19 no treinamento', 
                    xlabel = 'Epochs', ylabel='Acurácia')

VGG19_vacc.plot(ax=axs[1], title='Acurácia da estutura VGG19 na validação', 
                   xlabel = 'Epochs', ylabel='Acurácia')
    



```

## Comparações Gerais

```python
fig, axs =  plt.subplots(6, 2, figsize=(15,40), sharey=True)

SemTL_acc.plot(ax=axs[0][0], title='Acurácia da estutura sem Transfer Learning no treinamento', 
                    xlabel = 'Epochs', ylabel='Acurácia')

SemTL_vacc.plot(ax=axs[0][1], title='Acurácia da estutura sem Transfer Learning na validação', 
                   xlabel = 'Epochs', ylabel='Acurácia')

VGG16_acc.plot(ax=axs[1][0], title='Acurácia da estutura VGG16 no treinamento', 
                    xlabel = 'Epochs', ylabel='Acurácia')

VGG16_vacc.plot(ax=axs[1][1], title='Acurácia da estutura VGG16 na validação', 
                   xlabel = 'Epochs', ylabel='Acurácia')

InceptionResNetV2_acc.plot(ax=axs[2][0], title='Acurácia da estutura InceptionResNetV2 no treinamento', 
                    xlabel = 'Epochs', ylabel='Acurácia')

InceptionResNetV2_vacc.plot(ax=axs[2][1], title='Acurácia da estutura InceptionResNetV2 na validação', 
                   xlabel = 'Epochs', ylabel='Acurácia')

InceptionV3_acc.plot(ax=axs[3][0], title='Acurácia da estutura InceptionV3 no treinamento', 
                    xlabel = 'Epochs', ylabel='Acurácia')

InceptionV3_vacc.plot(ax=axs[3][1], title='Acurácia da estutura InceptionV3 na validação', 
                   xlabel = 'Epochs', ylabel='Acurácia')

Xception_acc.plot(ax=axs[4][0], title='Acurácia da estutura Xception no treinamento', 
                    xlabel = 'Epochs', ylabel='Acurácia')

Xception_vacc.plot(ax=axs[4][1], title='Acurácia da estutura Xception na validação', 
                   xlabel = 'Epochs', ylabel='Acurácia')

VGG19_acc.plot(ax=axs[5][0], title='Acurácia da estutura VGG19 no treinamento', 
                    xlabel = 'Epochs', ylabel='Acurácia')

VGG19_vacc.plot(ax=axs[5][1], title='Acurácia da estutura VGG19 na validação', 
                   xlabel = 'Epochs', ylabel='Acurácia')
    
    
    

```

```python
acc = pd.DataFrame({
    'VGG19':    VGG19_acc.mean(axis=1),
    'VGG16': VGG16_acc.mean(axis=1),
    'Xception': Xception_acc.mean(axis=1),
    'InceptionV3': InceptionV3_acc.mean(axis=1),
    'InceptionResNetV2': InceptionResNetV2_acc.mean(axis=1),
    'semTL': SemTL_acc.mean(axis=1)
    
                    })

vacc = pd.DataFrame({
    'VGG19':    VGG19_vacc.mean(axis=1),
    'VGG16': VGG16_vacc.mean(axis=1),
    'Xception': Xception_vacc.mean(axis=1),
    'InceptionV3': InceptionV3_vacc.mean(axis=1),
    'InceptionResNetV2': InceptionResNetV2_vacc.mean(axis=1),
    'semTL': SemTL_vacc.mean(axis=1)
    
                    })
#plt.figure(0, figsize=(15,5))

fig, axs = plt.subplots(nrows=1,ncols=2, figsize=(13,5), sharey=True)
acc.plot(ax=axs[0], legend = False, title= 'Média de acurácia nos treinamentos', xlabel='Epochs', ylabel='Acurracy')
vacc.plot(ax=axs[1], legend=False, title='Média de acurácia nas validações', xlabel='Epochs', ylabel='Accuracy')
fig.legend(acc.keys(),loc='center left', bbox_to_anchor=(.9,.7))



```
